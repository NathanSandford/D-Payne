```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

# Determine Successful and Unsuccessful Models

```python
khome='/global/scratch/nathan_sandford/kurucz_old'
list_models_done = [os.path.basename(x)[5:10] for x in glob.glob(khome+"/grids/at*/spec/*")]
list_models_done_prior = [os.path.basename(x)[5:10] for x in glob.glob(khome+"/Sync_Spectra_All_Atm_m0/*")]
list_models_done = np.apend(list_models_done,list_models_done_prior)

list_models_all = [os.path.basename(x) for x in glob.glob(khome+"/grids/?[!t]*")]
list_models_all = np. append(list_models_all,list_models_done_prior)

list_models_fail = [x for x in list_models_all if x not in list_models_done]

list_models_done.sort()
list_models_all.sort()
list_models_fail.sort()

fail = [list_models_all.index(x) for x in list_models_all if x not in list_models_done]
success = [list_models_all.index(x) for x in list_models_all if x in list_models_done] 
```

```python
temp = np.load('/global/home/users/nathan_sandford/kurucz/input/grid_fehm0.npz')
failed_labels = temp['labels'][:,fail]
successful_labels = temp['labels'][:,success]
```

```python
plt.scatter(failed_labels[0,:],failed_labels[1,:],label='failed',marker='.')
plt.scatter(successful_labels[0,:],successful_labels[1,:],label='successful',marker='.')
plt.gca().invert_xaxis()
plt.xlabel('T [k]')
plt.ylabel('log(g)')
plt.show()

plt.scatter(failed_labels[0,:],failed_labels[2,:],label='failed',marker='.')
plt.scatter(successful_labels[0,:],successful_labels[2,:],label='successful',marker='.')
plt.gca().invert_xaxis()
plt.xlabel('T [k]')
plt.ylabel('[Fe/H]')
plt.show()

plt.scatter(failed_labels[1,:],failed_labels[2,:],label='failed',marker='.')
plt.scatter(successful_labels[1,:],successful_labels[2,:],label='successful',marker='.')
plt.xlabel('log(g)')
plt.ylabel('[Fe/H]')
plt.show()
```
